# Sieve of Eratosthenes 

```julia
function primeSieve(n)
    a=zeros(Int64,10000)
    for i in 2:n
        if(a[i]==0)
            for j in (i*i):n
                a[j]=1
            end
        end
    end
    for i in 2:n
        if(a[i]==0)
            println(i)
        end
    end
end
```

```julia
primeSieve(100)
```

# Prime Factorisation 

```julia
function PrimeFactor(n)
    a=zeros(Int64,n)
    for i in 2:n
        a[i]=i
    end
    for i in 2:n
        if(a[i]==i)
            for j in (i*i):i:n
                if(a[j]==j)
                    a[j]=i
                end
            end
        end
    end
    while(n!=1)
        println(a[n])
        n=div(n,a[n])
    end
end
```

```julia
PrimeFactor(100)
```

# Tower of Hanoi 

```julia
function towerOfHanoi(n,source,destination,helper)
    if(n==0)
        return
    end
    towerOfHanoi(n-1,source,helper,destination)
    println("Move from $source to $dest ")
    towerOfHanoi(n-1,helper,destination,source)
end
```

```julia
towerOfHanoi(3,'a','b','c')
```

# 0-1 Knapsack

```julia
function Knapsack(weight[Int64],value[Int64],numberOfItems,capacity)
    if(numberOfItems==0 || capacity==0)
        return 0
    end
    if(weight[numberOfItems]>capacity)
        return Knapsack(weight,value,numberOfItems-1,capacity)
    end
    return max(Knapsack(weight,value,numberOfItems-1,capacity-weight[numberOfItems])+value(numberOfItems),Knapsack(weight,value,numberOfItems-1,capacity))
end
```

# Factorials 

```julia
function fact(n)
    if(n==0)
        return 1
    end
    if(n==1)
        return 1
    end
    return n*fact(n-1)
end
```

```julia
fact(10)
```

# GCD 

```julia
function GCD(a,b)
    if(a==0)
        return b
    end
    if(b==0)
        return a
    end
    if(a<b)
        a,b=b,a
    end
    ans=a%b
    while(ans!=0)
        a=b
        b=ans
        ans=a%b
    end
    return b
        
end
```

```julia
GCD(10,20)
```
